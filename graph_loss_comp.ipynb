```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv('loss_dict.csv') 
df = pd.read_csv('loss_dict_comp.csv')
print(df.describe())
df.head()
```

```python
# Filtro
start_epoch = 1
end_epoch = 150
sets = ["train", "valid"]

df = df[(df['epoch'] >= start_epoch) & (df['epoch'] <= end_epoch)]
df = df[df["set"].isin(sets)]
```

```python
# Our Loss
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='epoch', y='our_loss', hue='set')
plt.title('Our Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Our Loss')
plt.legend(title='Set')
plt.grid(True)
plt.show()
```

```python
# Their Loss
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='epoch', y='their_loss', hue='set')
plt.title('Their Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Their Loss')
plt.legend(title='Set')
plt.grid(True)
plt.show()
```

```python
# Acc
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='epoch', y='acc', hue='set')
plt.title('Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(title='Set')
plt.grid(True)
plt.show()
```
